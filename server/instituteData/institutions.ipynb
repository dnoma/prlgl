```python
import pandas as pd
```

```python
institutions = pd.read_csv("arbitral-dataset.csv")
```

```python
institutions.head()
```

```python
institutions.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1, inplace=True)
```

```python
institutions.head()
```

```python
institutions.sample(30)
```

```python
import json

# Define a function to remove double quotes from the start and end of a string, if present.
def remove_quotes(s):
    if isinstance(s, str):
        return s.strip('"')
    return s  # Return the value unchanged if it's not a string

# Apply this function to every element in the 'Institution' column
institutions['Flag'] = institutions['Flag'].apply(remove_quotes)

# Now convert the 'Institution' column to a list and then to a JSON list of strings
institutions_list = institutions['Flag'].dropna().tolist()
json_list = json.dumps(institutions_list)  # Convert list to JSON string

print(json_list)
```
